### Categorical Univariate analysis for customers without payment difficulties in logarithmic scale

```python
def cat_plot(df,col,title,hue =None):
    
    sns.set_style('whitegrid')
    sns.set_context('talk')
    plt.rcParams["axes.labelsize"] = 20
    plt.rcParams['axes.titlesize'] = 22
    plt.rcParams['axes.titlepad'] = 30
    
    
    temp = pd.Series(data = hue)
    fig, ax = plt.subplots()
    width = len(df[col].unique()) + 7 + 4*len(temp.unique())
    fig.set_size_inches(width , 8)
    plt.xticks(rotation=45)
    plt.yscale('log')
    plt.title(title)
    ax = sns.countplot(data = df, x= col, order=df[col].value_counts().index,hue = hue,palette='Paired') 
        
    plt.show()
```

### plotting income type

```python
cat_plot(target0,col='NAME_INCOME_TYPE',title='Income type distribution',hue='CODE_GENDER')
```
